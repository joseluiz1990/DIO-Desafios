```python
def criar_usuario (lista_usuarios):

    opcao = input("Criar usuario (S/N)? ")
    while opcao == "S":
      cpf = input("Digite CPF (só numeros, sem pontos e sem traço): ")
      for chave_cpf in lista_usuarios:
        if cpf == chave_cpf:
          cadastrado = True
          break
        else:
          cadastrado = False

      if cadastrado == False:
        # cria usuario
        nome = input("Digite nome: ")
        data_nascimento = input("Digite data de nascimento: ")
        logradouro = input("Digite Rua/Av: ")
        numero = input("Digite numero: ")
        bairro = input("Digite bairro: ")
        cidade = input("Digite cidade: ")
        estado = input("Digite estado: ")
        """endereco = {
                    "logradouro": logradouro,
                    "numero": numero,
                    "bairro": bairro,
                    "cidade": cidade,
                    "estado": estado
                    }"""
        usuario = {
                    "cpf": cpf,
                    "nome": nome,
                    "data_nascimento": data_nascimento,
                    #"endereco": endereco
                    "logradouro": logradouro,
                    "numero": numero,
                    "bairro": bairro,
                    "cidade": cidade,
                    "estado": estado
                    }
        lista_usuarios[cpf]=usuario
        print("Correntista cadastrado com sucesso!")
      else:
          print("CPF já cadastrado- EXISTENTE!")
      # if end
      opcao = input("Criar usuario (S/N)? ")
    # while end
    return lista_usuarios
#----------------------------------------------------------
def criar_conta(lista_usuarios, lista_contas):
    opcao = input("Criar conta-corrente (S/N)? ")
    while opcao == "S":
      cpf = input("Digite CPF (só numeros, sem pontos e sem traço): ")

      for chave_cpf in lista_usuarios:
        if cpf == chave_cpf:
          cadastrado = True
          break
        else:
          cadastrado = False

      if cadastrado == True:
        # cria usuario
          len(lista_contas) + 1
          numero_conta = str(len(lista_contas) + 1)
          usuario = cpf
          #usuario = lista_usuarios[cpf]
          conta_corrente = {
                            "conta": numero_conta,
                            "agencia": "0001",
                            "usuario": usuario
                           }
          lista_contas[numero_conta] = conta_corrente
          print("Conta criada com sucesso!")
          print(conta_corrente)
      else:
          print("CPF não cadastrado! Crie usuario!")
      #if end
      opcao = input("Criar conta-corrente (S/N)? ")
      # while end
    return lista_contas
#----------------------------------------------------------
# positional only
# positional antes da barra
def depositar(saldo, valor, extrato,/):
  valor = float(input("Informe valor do depósito: "))
  if valor > 0:
    saldo += valor
    extrato += f"Depósito: R$ {valor:.2f}\n"
    print("Depósito realizado com sucesso!")
  else:
    print("Deposito falhou! Valor negativo inválido.")
  return saldo, extrato
#----------------------------------------------------------
# keyword only
# keyword depois do asterisco
def sacar(*, saldo, valor, extrato, limite, numero_saques, limite_saques):
  if numero_saques < limite_saques:
    valor = float(input("Informe valor do saque: "))
    if valor <= saldo:
      if valor <= limite:
        if valor > 0:
          saldo -= valor
          extrato += f"Saque: R$ {valor:.2f}\n"
          numero_saques = numero_saques + 1
          print("Saque realizado com sucesso!")
        else:
          print("Saque falhou! Valor negativo inválido.")
      else:
        print("Saque falhou! Saque maior que limite.")
    else:
      print("Saque falhou! Saldo insuficiente.")
  else:
    print("Saque falhou! Quantidade de saques excedido.")
  return saldo, extrato, numero_saques
#----------------------------------------------------------
# positional => saldo; keyword => extrato
# positional antes da barra
# keyword depois do asterisco
def extratificar(saldo, /, *, extrato):
  print(f"EXTRATO\n{extrato}")
  print(f"Saldo: R$ {saldo:.2f}")
#----------------------------------------------------------
# PRINCIPAL
#----------------------------------------------------------
plista_usuarios={'00000000000': {'cpf': '00000000000', 'nome': '', 'data_nascimento': '', 'logradouro': '', 'numero': '', 'bairro': '', 'cidade': '', 'estado': ''}}
plista_contas={}
#----------------------------------------------------------
pvalor=0.0
psaldo = 0.0
plimite = 500.0
pextrato = ""
pnumero_saques = 0
plimite_saques = 3
menu = """
Digite operação desejada:
[d] Depositar
[s] Sacar
[e] Extrato
[u] Criar usuario
[c] Criar conta
[x] Sair
=> """

opcao = input(menu)

while opcao != "x":
  if opcao == "d":
    print(f"saldo antes = {psaldo}")
    print(f"extrato antes =\n {pextrato}")
    deposito=depositar(psaldo, pvalor, pextrato)
    psaldo = deposito[0]
    pextrato = deposito[1]
    print(f"saldo depois = {psaldo}")
    print(f"extrato depois =\n {pextrato}")

  elif opcao == "s":
    print(f"saldo antes = {psaldo}")
    print(f"extrato antes =\n {pextrato}")
    saque=sacar(saldo=psaldo, valor=pvalor, extrato=pextrato, limite=plimite, numero_saques=pnumero_saques, limite_saques=plimite_saques)
    psaldo = saque[0]
    pextrato = saque[1]
    pnumero_saques = saque[2]
    print(f"saldo depois = {psaldo}")
    print(f"extrato depois =\n {pextrato}")
    print(f"numero de saques = {pnumero_saques}")

  elif opcao == "e":
    extratificar(psaldo, extrato=pextrato)

  elif opcao == "u":
    plista_usuarios = criar_usuario(plista_usuarios)
    for usuario in plista_usuarios:
      print(plista_usuarios[usuario])

  elif opcao == "c":
    plista_contas = criar_conta(plista_usuarios, plista_contas)
    for conta in plista_contas:
      print(plista_contas[conta])

  opcao = input(menu)

else:
  print("Operação finalizada")
```
